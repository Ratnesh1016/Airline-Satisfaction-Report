# AIRLINE PASSENGER SATISFACTION PROJECT

## Exploratory Data Analysis

```python
# ðŸ“Œ Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# âœ… Load dataset â€” replace this with your actual CSV file path
df = pd.read_csv("C:\\Users\\ratne\\Downloads\\test.csv.zip")

# âœ… Preview the first 5 records
df.head()
```

## Data Overview and Summary

```python
# âœ… Basic info about dataset
df.info()

# âœ… Check for missing values
print(df.isnull().sum())

# âœ… Descriptive statistics
df.describe()
```

## Satisfaction Value Counts

```python
# âœ… Satisfaction distribution counts
df['satisfaction'].value_counts()
```

## Pie Chart â€” Satisfaction Distribution

```python
# âœ… Pie chart for passenger satisfaction distribution
labels = df['satisfaction'].value_counts().index
sizes = df['satisfaction'].value_counts().values

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff6347','#1e90ff'])
plt.title('Passenger Satisfaction Distribution')
plt.show()
```

## Satisfaction by Travel Class

```python
# âœ… Countplot of Satisfaction by Travel Class
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Class', hue='satisfaction', palette='Set2')
plt.title('Satisfaction by Travel Class')
plt.show()
```

## Average Service Ratings by Satisfaction

```python
# âœ… Barplot of average service ratings by satisfaction group
services = ['Inflight wifi service', 'Seat comfort', 'Food and drink', 
            'Cleanliness', 'Inflight entertainment']

plt.figure(figsize=(10,6))
df.groupby('satisfaction')[services].mean().T.plot(kind='bar')
plt.title('Average Service Ratings by Satisfaction')
plt.ylabel('Average Rating')
plt.xlabel('Service Feature')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## Flight Distance vs Satisfaction

```python
# âœ… Boxplot of Flight Distance by Satisfaction
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='satisfaction', y='Flight Distance', hue='satisfaction', palette='pastel', legend=False)
plt.title('Flight Distance vs Satisfaction')
plt.show()
```

## Correlation Heatmap

```python
# âœ… Correlation heatmap of numerical features
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```
